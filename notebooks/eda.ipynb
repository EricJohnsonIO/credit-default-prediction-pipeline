```python
from src.preprocessing import FeatureEngineer

print("Timeing")
fe = FeatureEngineer()
x_train_eng = fe.fit_transform(x_train.copy())
x_train_eng["target"] = y_train.values

eng_df = x_train_eng.copy()

print("timeing done")
```

```python
plt.figure(figsize=(18,14))
sns.heatmap(eng_df.corr(numeric_only=True), annot=True,cmap="coolwarm",vmin=-1,vmax=1)
```

```python
raw_corr = train_df.corr(numeric_only=True)["default.payment.next.month"].drop("default.payment.next.month").abs().sort_values()
eng_corr = eng_df.corr(numeric_only=True)["target"].drop("target").abs().sort_values()

fig, axes = plt.subplots(1, 2, figsize=(18, 7))
sns.barplot(x=raw_corr.values, y=raw_corr.index, ax=axes[0], color="steelblue")
sns.barplot(x=eng_corr.values, y=eng_corr.index, ax=axes[1], color="coral")
axes[0].set_title("Raw Features")
axes[1].set_title("Engineered Features")
plt.tight_layout()
plt.show()
```

```python
plot_feats = ["DELAY_RISK", "DELAY_RECENT_WORST", "DELAY_MAX", "DELAY_RECENT",
              "LIMIT_BAL", "PAY_AMT_MEAN"]


fig, axes = plt.subplots(2, 3, figsize=(18, 10))
axes = axes.flatten()

for i, feat in enumerate(plot_feats):
    all_vals = eng_df[feat]
    p1, p99 = all_vals.quantile(0.01), all_vals .quantile(0.99)
    for label, color in [(0, "steelblue"), (1, "tomato")]:
        subset = eng_df[eng_df["target"] == label][feat].clip(p1, p99)
        axes[i].hist(subset, bins=40, alpha=0.5, color=color,
                     label=f"{'Default' if label else 'No Default'}", density=True)
    axes[i].set_title(feat)
    axes[i].legend()

plt.tight_layout()
plt.show()
```
